```python
def getIPEntriesFromOTXAlienVault(df):
    
    final_df = pd.DataFrame()
    api_key = '766db8000d50178e5cf3b6ce9b841811ca9249aa865d5f4c7aa14d241b4656af'
    useful_columns = df.columns
    otx_file_usefulColumns = ['otx_id','otx_name','otx_description', 'otx_tags', 'otx_attack_ids', 'otx_malware_families', 'otx_adversary', 'otx_modified', 'otx_created', 'otx_references', 'otx_author.username','otx_targeted_countries' ]

    for i in range(len(df)):
        
        print(i)
        
        ioc_type = df.at[i,'ioc_type']
        el = pd.DataFrame(df[df.index == i])

        
        if(ioc_type == 'ip:port'):
            ioc = df.at[i,'ioc'].split(':')[0]
            otxAlienVault_df = searchIpv4inOTXAlienVault(ioc)
            
        if(ioc_type == 'url'):
            ioc = df.at[i,'ioc']
            otxAlienVault_df = searchURLinOTXAlienVault(ioc)

            
        if(ioc_type == 'domain'):
            ioc = df.at[i,'ioc']
            otxAlienVault_df = searchDomaininOTXAlienVault(ioc)

        if(ioc_type == 'md5_hash'):
            ioc = df.at[i,'ioc']
            otxAlienVault_df = searchFileinOTXAlienVault(ioc)

        if(ioc_type == 'sha256_hash'):
            ioc = df.at[i,'ioc']
            otxAlienVault_df = searchFileinOTXAlienVault(ioc)

        if(otxAlienVault_df.empty == False):

            otxAlienVault_df = pd.DataFrame.from_dict(pd.json_normalize(otxAlienVault_df[0])).add_prefix('otx_')

            rows_otx = pd.DataFrame()
            rows_otx = otxAlienVault_df[otx_file_usefulColumns]
            rows_threats = pd.DataFrame()
            rows_threats = el[useful_columns]

            rows = pd.concat([rows_threats,rows_otx], axis = 1)

            rows[useful_columns] = rows[useful_columns].fillna(method='ffill')

            if(len(final_df) > 0):
                rows = rows.iloc[1:, :]

            final_df = pd.concat([final_df,rows], ignore_index = True)

            
    return final_df
```

```python
def searchIpv4inOTXAlienVault(ip):
    api_key = '766db8000d50178e5cf3b6ce9b841811ca9249aa865d5f4c7aa14d241b4656af'
    otx_url = 'https://otx.alienvault.com/api/v1/indicators/IPv4/'+ ip +'/general'
    current_pulse = requests.get(otx_url,headers={"X-OTX-API-KEY":api_key})
    otxAlienVault_df = pd.DataFrame()
    
    if(str(current_pulse) == '<Response [200]>' ):                 
        current_pulse = json.loads(current_pulse.content.decode("utf-8", "ignore"))

        tmp = pd.DataFrame.from_dict(pd.json_normalize(current_pulse)).reset_index(drop = True)        
        otxAlienVault_df = pd.DataFrame.from_dict(pd.json_normalize(tmp['pulse_info.pulses'])).T
        
    return otxAlienVault_df   
```

## get entries in OTX Alien Vault

```python
def search(ip):
    
    max_retries = 10 
    retry_delay = 20
    
    otx_url = 'https://otx.alienvault.com/api/v1/indicators/url/'+ ip + '/general'    
    attempts = 0
        
    while attempts < max_retries:
        
        try:
            current_pulse = requests.get(otx_url,headers={"X-OTX-API-KEY":api_key})
            
            #print(f" Http request for { otx_url } url")
            
            if(str(current_pulse) == '<Response [200]>' ):
                current_pulse = json.loads(current_pulse.content.decode("utf-8", "ignore"))
                tmp = pd.DataFrame.from_dict(pd.json_normalize(current_pulse)).reset_index(drop=True)

                otxAlienVault_df = pd.DataFrame()

                if 'pulse_info.pulses' in tmp.columns:

                    details = pd.DataFrame.from_dict(pd.json_normalize(tmp['pulse_info.pulses'])).T

                    if(details.empty == False):

                        otxAlienVault_df = pd.DataFrame.from_dict(pd.json_normalize(details[0])).add_prefix('otx_')

                        tmp = tmp.drop(['pulse_info.pulses'], axis=1)

                        concatenated_df = pd.DataFrame()

                        concatenated_df = pd.concat([tmp, otxAlienVault_df], axis=1)
                        
                        print(f" Found! ")

                        return concatenated_df
                    
            return pd.DataFrame()    
                    
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON for URL {ip}: {e}")
            return pd.DataFrame()
        
        except requests.exceptions.HTTPError as errh:
            print(f"HTTP Error: {errh}")
            
        except requests.exceptions.ConnectionError as errc:
            print(f"Error Connecting: {errc}")
            
        except requests.exceptions.Timeout as errt:
            print(f"Timeout Error: {errt}")
            
        except requests.exceptions.RequestException as err:
            print(f"Request Exception: {err}")
            
        attempts += 1
        print(f"Retrying in {retry_delay} seconds...")
        time.sleep(retry_delay)
        
    return pd.DataFrame()    
```

```python
def build_dataframe(df):
    
    dataframe = pd.DataFrame()
    
    for i in range(len(df)):
                    
            print(i)
            
            url = df[i]
            
            tmp_row = search(url)
            
            if(tmp_row.empty == False):
                        
                dataframe = pd.concat([dataframe, tmp_row])
            
    return dataframe
```

# Entries

```python
entries = pd.read_csv(path_Entries, low_memory = False)
entries
```

## Data Cleaning

#### Data type Conversion

- analysis_date field : from string to date
- first_seen field : from string to date
- last_seen field : from string to date

```python
entries['first_seen'] = pd.to_datetime(entries['first_seen'])
entries['last_seen'] = pd.to_datetime(entries['last_seen'])
entries['analysis_date'] = pd.to_datetime(entries['analysis_date'])
```

#### Missing Values

- replace NaN in 'delivery_method' with 'unknown'

```python
entries['delivery_method'] = entries['delivery_method'].replace(np.nan, 'unknown')
```

```python
entries.dropna(subset=['sha1', 'sha256', 'md5'], inplace=True)
entries.reset_index(drop=True, inplace=True)

entries
```

```python
def analyse_columns(df):
    
    columns = df.columns
    for i in range(len(columns)):
        
        compute_completeness(df, columns[i])

def compute_completeness(df, col):
    completeness = 1 - df[col].isna().mean()
    print(f"Completeness for '{(col)}': {completeness * 100:.2f}%")
```

```python
analyse_columns(entries)
```

# Threat Vendor Analysis

```python
tv_analysis = pd.read_csv(path_TV_Analysis, low_memory = False)
tv_analysis.head(3)
```

## Data Cleaning

#### Missing Values

- drop rows with missing values related to sha1, sha256, md5 and threat_vendor_name field because they cannot be associate to any file

```python
tv_analysis.dropna(subset=['sha1', 'sha256', 'md5', 'threat_vendor_name'], inplace=True)
tv_analysis
```

- drop row if both the fields 'verdict' and 'malware' are missing because any information can be taken from that row, so it will be useless

```python
tv_analysis.dropna(subset=['malware', 'verdict'], how='all', inplace=True)
tv_analysis.reset_index(drop=True, inplace=True)
tv_analysis
```

#### Standardization 

- in the field 'verdict': the terms 'malware infection' and 'malware detected' are associated to 'malicious'

```python
tv_analysis['verdict'].unique()
```

```python
tv_analysis['verdict'] = tv_analysis['verdict'].replace('malware infection', 'malicious').replace('malware detected', 'malicious')
```

```python
tv_analysis['verdict'].unique()
```

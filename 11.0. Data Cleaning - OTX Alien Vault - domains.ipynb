# Data Deduplication

```python
def find_duplicates(input_list):
    seen = {}
    duplicates = []
    for item in input_list:
        if item in seen:
            duplicates.append(item)
        else:
            seen[item] = 1
    return duplicates
```

## Field analyzed: 'hash'

```python
df['hash'].unique()
```

```python
files = df[df['hash'].notna()]
files 
```

```python
dup_files = find_duplicates(files.hash)
len(dup_files)
```

```python
# By a first look at the duplicates in the DB, it seems that there are 
# multiple ip addresses related to the same hash with all the other fields in common

# This function will return true if this assumption will be correct

def analyze_duplicates(df, duplicates):
    
    hp = True
    x = []
    
    for i in range(len(duplicates)):
        
        el = duplicates[i]
        
        dataframe = df[df['hash'] == el]
        dataframe.reset_index(drop=True, inplace=True)
                        
        tmp_avast      = str(dataframe.at[0, 'detections.avast'])
        tmp_clamav     = str(dataframe.at[0, 'detections.clamav'])
        tmp_msdefender = str(dataframe.at[0, 'detections.msdefender'])
        tmp_date       = str(dataframe.at[0, 'date'])
                
        for j in range(len(dataframe) - 1):
            
            if(str(dataframe.at[j + 1 , 'date']) != tmp_date ):
                print(f"{dataframe.at[j + 1 , 'date']} and {tmp_date} are different")
                
                hp = False
                
            if(str(dataframe.at[j + 1 , 'detections.msdefender']) != tmp_msdefender ):
                print(f"{dataframe.at[j + 1 , 'detections.msdefender']} and {tmp_msdefender} are different")

                hp = False
                
            if(str(dataframe.at[j + 1 , 'detections.avast']) != tmp_avast ):
                print(f"{dataframe.at[j + 1 , 'detections.avast']} and {tmp_avast} are different")

                hp = False
                
            if(str(dataframe.at[j + 1 , 'detections.clamav']) != tmp_clamav ):
                print(f"{dataframe.at[j + 1 , 'detections.clamav']} and {tmp_clamav} are different")

                hp = False
            
    return hp
    
    
```

```python
print(f" The previous hypothesys was correct?  {analyze_duplicates(df, dup_files)}")
```

```python
headers = { 'API-KEY': 'a4582c8cb2e30fc6b444b509bb6a9a56' }

data = {
    'query': 'get_iocs',
    'days' : 7
}

pool = urllib3.HTTPSConnectionPool('threatfox-api.abuse.ch', port=443, maxsize=500)

json_data = json.dumps(data)
response = pool.request("POST", "/api/v1/", body=json_data)
response = response.data.decode("utf-8", "ignore")
json_response =  json.loads(response)


entries = pd.DataFrame.from_dict(pd.json_normalize(json_response))
iocs_ThreatFox_df = pd.DataFrame.from_dict(pd.json_normalize((pd.DataFrame.from_dict(pd.json_normalize(entries['data'])).T)[0]))
iocs_ThreatFox_df.columns
```

```python
iocs_ThreatFox_df
```

```python
columns = ['ioc', 'ioc_type', 'threat_type', 'sha256_hash', 'sha3_384_hash',
       'sha1_hash', 'md5_hash', 'source', 'first_seen', 'last_seen',
       'file_name', 'file_type', 'reporter','threat',
       'origin_country', 'malware','tags', 'maliciousness', 'status', 'details', 'references']
```

```python
export = pd.DataFrame()
# fields = iocs_ThreatFox_df.columns
export[columns] = ''
iocs_ThreatFox_df.columns
```

```python
export.columns
```

```python
export[[ 'ioc','first_seen', 'last_seen','threat_type','ioc_type','threat','references','reporter','tags']] =  iocs_ThreatFox_df[[ 'ioc','first_seen', 'last_seen','threat_type','ioc_type','malware','reference','reporter','tags']]

export['source'] = 'Threat Fox'

export
```

```python
for i in range(len(export)):
    
    values = str(iocs_ThreatFox_df.at[i,'malware_alias']).split(',')
    
    for j in range(len(values)):
        values[j] = values[j].lower()
    
    if((str(iocs_ThreatFox_df.at[i,'malware_printable']) != 'None' ) & (str(iocs_ThreatFox_df.at[i,'malware_alias']) != 'None' )):
        el = iocs_ThreatFox_df.at[i,'malware_printable'].lower()
        
        if(el not in values):
            val = values + [el]
            val = json.dumps(val)

            export.loc[i,'malware'] = val
    
export
```

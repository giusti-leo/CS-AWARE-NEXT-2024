## FOLLOWING THE INTEGRATION SCHEMA PERFORMED ON PAPER

Final dataset has the following fields:

- ENTRIES: final_df ( // ID_ENTRY , source // , 'ioc', 'threat_type', 'ioc_type', 'malware', 'reference', 'first_seen', 'last_seen', 'reporter',  )
       
- TAGS:  final_tags_df( // ID_ENTRY, source , tag_name // )

- ALIAS: final_MALWARES_df ( // malware // , alias )

- MALWARES: final_MALWARES_df ( // malware //)


```python
final_ENTRIES_df = pd.DataFrame()

final_ENTRIES_df = pd.concat([entries_FT, entries_TF, entries_UH], ignore_index= True)

final_ENTRIES_df
```

## TAGS

```python
final_TAGS_df = pd.DataFrame()

final_TAGS_df = pd.concat([tags_UH, tags_TF], ignore_index= True)

final_TAGS_df
```

### MALWARES

```python
final_MALWARES_df = pd.DataFrame()

final_MALWARES_df = pd.concat([malwares_FT, malwares_TF], ignore_index= True)

final_MALWARES_df
```

### ALIAS

```python
final_ALIAS_df = pd.DataFrame()

final_ALIAS_df = pd.concat([alias_TF], ignore_index= True)

final_ALIAS_df
```

# SAVE into ..\IocSegnalations\PostDataIntegration\ folder

```python
dataframes_dict = {}

# Add the DataFrames to the dictionary with keys
dataframes_dict['Entries'] = final_ENTRIES_df
dataframes_dict['Tags'] = final_TAGS_df
dataframes_dict['Malwares'] = final_MALWARES_df
dataframes_dict['Alias'] = final_ALIAS_df
```

```python
save_directory = "C:\\Users\\leona\\OneDrive\\Desktop\\Tesi\\Pipeline\\Datasets\\Pipeline\\IocSegnalations\\PostDataIntegration\\"
for key, df in dataframes_dict.items():
    file_name = f"{key}.csv"
    file_path = save_directory + file_name
    df.to_csv(file_path, index=False)
```
